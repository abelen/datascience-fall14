```python
# Find any potential outliers in our data,
# first we check any price difference outliers
price = dataset[:,1]
std = price.std()
avg = np.average(price)
new_max = np.max(price)
new_min = np.min(price)
low_cutoff = avg - (3 * std)
high_cutoff = avg + (3 * std)
print 'The average price difference is ', avg
print 'The low cutoff is ', low_cutoff
print 'The high cutoff is ', high_cutoff
```

```python
# Remove any price that is outside 3 standard deviations
# away from mean
new_price_set = np.array([])
new_target = np.array([])
for i in range(len(price)):
    if (price[i] <= high_cutoff and price[i] >= low_cutoff):
        new_price_set = np.append(new_price_set,price[i])
        new_target = np.append(new_target, target[i])
```
