```python
#Exercise 3.2
plot_y_values = [];
num_of_features = boston.data.shape[1]
best_features = [];
    
# iterate through all the different percentages of data sets,
# find the number of features that have the lowest squared error
# and put it's
for percentages in boston_list_tuples:
    training_size = int (boston_size * percentages[0]) + 1
    test_size = int(boston_size * percentages[1])
    
    boston_x_train = diabetes.data[0:training_size-1]
    boston_x_test = diabetes.data[training_size:(training_size + test_size -1)]
    
    boston_y_train = diabetes.target[0:training_size-1]
    boston_y_test = diabetes.target[training_size:(training_size + test_size -1)]
    
    # set up sentinel values for the lowest feature count and
    # the mean squared error
    lowest_feature_count = 1000
    lowest_feature_value = float("inf")
    for i in range(1,num_of_features + 1):
        
        rfe = RFE(estimator=regr, n_features_to_select=i, step=1)
        rfe.fit(boston_x_train, boston_y_train)
        current_val = np.mean((rfe.predict(boston_x_test) - boston_y_test) ** 2)

        # check if the current number of features' squared is lower than the current
        # one
        if (current_val < lowest_feature_value):
            lowest_feature_count = i;
            lowest_feature_value = current_val;
    
    #update the number of 
    best_features.append(lowest_feature_count)
    plot_y_values.append(lowest_feature_value)

label_list = ["20% training 80% test", "40% training 60% test",
              "60% training 40% test", "80% training 20% test"]

for i in range(0,4):
    print label_list[i], " number of features best used ", best_features[i]
    
```
