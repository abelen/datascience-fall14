```python
#Assignment Part 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
# Loading the provided three CSV files
players = pd.read_csv('players.csv')
countries = pd.read_csv('countries.csv')
events = pd.read_csv('events.csv')
```

```python
# Count the total number of players whose names start with a Vowel
players['name'].str.extract('(^[AEOIU])').count()
```

```python
# Find all players within the USA, that start with A.
a_player = players[(players['country_id'] == 'USA') & (players['name'].str.extract( '^([A])') )];
a_player
```

```python
# Constructing a dataframe that takes the players and joins them
# with the countries and then does a count on the groupby
# country_name

players_countries = pd.merge(players, countries, on='country_id')
players_countries.groupby('name_y').count()['name_y']
```

```python
# Left outer join query with countries and the players,
# which then counts up the number of players that were born in 1975
# and 0 for any countries that have no players born in 1975
left_outer1 = pd.merge(countries, players, how='left', on='country_id', sort='birthdate');
left_outer1 = left_outer1[(left_outer1['birthdate'] >= '1975') & (left_outer1['birthdate'] < '1976')]
#print type(left_outer1)
left_outer1 = left_outer1[['name_x', 'birthdate']].groupby('name_x')['birthdate'].count()

left_outer2 = pd.merge(countries, players, how='left', on='country_id', sort='birthdate');
left_outer2 = left_outer2[['name_x']]
left_outer2['0'] = 0
left_outer2 = left_outer2[['name_x', '0']].groupby('name_x')['0'].sum()

new_left = left_outer1.append(left_outer2)
df = pd.DataFrame(new_left)
df.icol(0).index
new_val = df.groupby(df.icol(0).index).sum()
print new_val
```
